# Task 1.3 — What the Paper Claims to Improve

## Main baseline / prior method

**Baseline method:** **SVM-RFE (Support Vector Machine Recursive Feature Elimination)** by Guyon et al. (2002).

---

## Limitation identified in the baseline

SVM-RFE is **monotonic**: once a feature is eliminated, it cannot come back.

That becomes risky when the early SVM model is noisy (common with many noisy features). A useful feature can be removed early due to a wrong importance estimate, and then it is permanently lost.

**Paper reference:** Introduction, the paragraph explaining the “monotonic” property and citing Xu et al. (2009).

---

## How the paper’s method tries to overcome it

FGM is **non-monotonic**. It generates multiple sparse feature subsets over iterations, adds them to a working set, and combines them via an MKL-style optimization.

Since each new subset is generated from scratch using the current dual variables, a feature that was missing earlier can still appear later. This lets the method correct earlier mistakes.

**Paper reference:** Section 2.2, Algorithm 1 (cutting-plane loop).

---

## One scenario where FGM may not outperform SVM-RFE

If the dataset is relatively clean (few noisy features) and you are not forcing extreme sparsity (\(B\) not very small), then SVM-RFE is less likely to eliminate an important feature early.

In that setting, SVM-RFE can be competitive and faster, while FGM’s extra cutting-plane + MKL steps may not give a big accuracy gain.

**Paper connection:** The paper says the monotonic issue becomes “extremely critical” mainly with many noisy features, and the synthetic results show the gap grows as noise increases (Section 3.3, Figure 2).
