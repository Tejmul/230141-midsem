# Task 1.2 — Key Assumptions

## Assumption 1

**Assumption:** Swapping the order of \(\min_d\) and \(\max_\alpha\) (minimax swap) gives a convex relaxation that stays close to the original sparse SVM objective.

**Why the method needs it:** FGM solves the relaxed QCQP/MKL form. If the relaxation is loose, the algorithm may converge but still select the wrong features for the true mixed-integer objective.

**Violation scenario:** If \(B\) is very small and informative features are strongly correlated, the relaxation can behave differently from the original combinatorial problem.

**Paper reference:** Section 2.1 (Convex Relaxation), Eq. (2) → Eq. (3) → Eq. (4).

---

## Assumption 2

**Assumption:** The most violated mask \(\hat{d}\) can be found by computing \(c_j = \sum_i \alpha_i y_i x_{ij}\), sorting \(c_j^2\), and selecting the top \(B\) features.

**Why the method needs it:** This is the feature-generation step. It keeps each cutting-plane iteration fast because it avoids solving a hard integer program.

**Violation scenario:** With interaction-heavy structure (e.g., XOR), individual features can look weak even when pairs/groups are informative, so selecting by \(c_j^2\) can miss the true useful subset.

**Paper reference:** Section 2.4 (Finding the Most Violated \(\hat{d}\)).

---

## Assumption 3

**Assumption:** The cutting-plane loop converges in a small number of iterations in practice (the paper reports around 10 is usually enough).

**Why the method needs it:** Each iteration solves an MKL subproblem. The method’s scalability depends on needing only a few rounds.

**Violation scenario:** If many different sparse subsets are nearly tied (flat objective), the method may keep adding subsets with tiny improvements and take many more iterations.

**Paper reference:** Section 2.7 (Computational Complexity): “Empirically, a maximum of 10 iterations is enough for FGM to converge.”
